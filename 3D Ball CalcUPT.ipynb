```python
def rk4b(y,t,h): # h being the interval step
    k1=np.zeros(8) # This is borrowed from lecture.
    k2=np.zeros(8)
    k3=np.zeros(8)
    k4=np.zeros(8)
    
    
    F3=f3(t,y)
    
    k1 = h*f3(t, y)
    k2 = h*f3(t + (1.0/2.0)*h, y + (1.0/2.0)*k1)
    k3 = h*f3(t + (1.0/2.0)*h, y + (1.0/2.0)*k2)
    k4 = h*f3(t + h, y + k3)
    
    y = y +  (k1 + 2 * k2 + 2 * k3 + k4)/6.0
    
    return y
```
