### We will be plotting the specified function for n = 11, 21, and 31, in the interval of $-\pi,\pi$
$$ f(x)= \pi /2 + 2*(sin(x)+sin(3x)/3+sin(5x/5)+....sin(nx/n)) $$


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x,n):
    
    f = np.pi/2 # Initialized the 0th term.  
    
    for i in range(n):
        
        f += 2*np.sin((2*i+1)*x)/(2*i+1)
        
    # This generates odd values for each term.
    
    y = f
    plt.plot(x,y,label ='n = '+str(2*n+1)) # Plots the functions
    return f

x = np.arange(-np.pi, np.pi, 0.001) #Generate the x values for the function.


# These are the nth term or the (2*n+1) value
f(x,5)
f(x,10)
f(x,15)

plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```
